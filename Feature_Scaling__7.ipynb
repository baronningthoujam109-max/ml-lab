```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/Dataset_07.csv") # Make sure this file is uploaded to your Colab environment or the path is correct
df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train.iloc[:, 4:] = sc.fit_transform(X_train.iloc[:, 4:])

X_test.iloc[:, 4:] = sc.transform(X_test.iloc[:, 4:])
```
